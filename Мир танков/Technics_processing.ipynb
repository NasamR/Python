```python
import numpy as np
import pandas as pd

pd.set_option('display.max_columns', None)
```

```python
df = pd.read_excel('Techincs.xlsx')
df.head()
```

```python
def split(x):
    x = x.split(' / ')
    x[-1] = x[-1].split(' ')[0]
    if len(x) < 3:
        for i in range(3 - len(x)):
            x.append(np.nan)
    return x

df['Урон'] = df['Урон'].apply(split)
df['Урон 1'] = df['Урон'].apply(lambda x: x[0])
df['Урон 2'] = df['Урон'].apply(lambda x: x[1])
df['Урон 3'] = df['Урон'].apply(lambda x: x[2])
```
